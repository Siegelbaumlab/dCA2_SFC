```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import csv
import warnings
warnings.filterwarnings('ignore')
```

```python
with open('Fig4_Density_Cre-2.csv','r') as file:
    reader = csv.reader(file)
    data = list(reader)
    
x_p = []
y_p = []

x_g = []
y_g = []

x_p_grey = []
y_p_grey = []

x_g_grey = []
y_g_grey = []

for el in data:
    if (el[2] == 'purple'):
        x_p.append(int(el[0]))
        x_p_grey.append(int(el[0]))
        y_p.append(float(el[4]))
        y_p_grey.append(float(el[3]))
    else:
        x_g.append(int(el[0]))
        x_g_grey.append(int(el[0]))
        y_g.append(float(el[4]))
        y_g_grey.append(float(el[3]))

data_p = []
for i in range(len(y_p)):
    d = int(y_p[i]*1000)
    for j in range(d):
        data_p.append(x_p[i])
        
data_p_grey = []
for i in range(len(y_p_grey)):
    d = int(y_p_grey[i]*1000)
    for j in range(d):
        data_p_grey.append(x_p_grey[i])
        
data_g = []
for i in range(len(y_g)):
    d = int(y_g[i]*1000)
    for j in range(d):
        data_g.append(x_g[i])
        
data_g_grey = []
for i in range(len(y_g_grey)):
    d = int(y_g_grey[i]*1000)
    for j in range(d):
        data_g_grey.append(x_g_grey[i])
        
prod_g = sum(y_g)
prod_g_grey = sum(y_g_grey)
prod_p = sum(y_p)
prod_p_grey = sum(y_p_grey)

y_g = [el/prod_g for el in y_g]
y_g_grey = [el/prod_g_grey for el in y_g_grey]
y_p = [el/prod_p for el in y_p]
y_p_grey = [el/prod_p_grey for el in y_p_grey]
```

```python
def width(i):
    return i-0
def width_o(i):
    return i+0

ax = plt.gca()

sns.kdeplot(data_p,color='grey',bw_adjust=1.6,ax=ax,lw=2,linestyle='--')
sns.kdeplot(data_p,color='#680aab',bw_adjust=1.6,ax=ax,lw=5)
sns.kdeplot(data_p_grey,color='#680aab',bw_adjust=1.6,ax=ax,lw=5,linestyle='--')

#plt.bar(list(map(width,x_p)),y_p,edgecolor="#680aab",color="white",alpha=0.7,width=0.8,linewidth=3)
#plt.bar(list(map(width,x_p_grey)),y_p_grey,edgecolor="#680aab",color="grey",alpha=0.7,width=0.8,linewidth=3,linestyle='--', bottom=y_p)


plt.bar(list(map(width,x_p_grey)),y_p_grey,edgecolor="#680aab",color="grey",alpha=0.7,width=0.8,linewidth=3,linestyle='--')
plt.bar(list(map(width,x_p)),y_p,edgecolor="#680aab",color="white",alpha=0.7,width=0.8,linewidth=3, bottom=y_p_grey)


#ax.set_xlabel("Number of Approach-avoidance \n events",fontsize=26, fontname='Arial')
ax.set_ylabel('', fontsize = 26, fontname='Arial')

ax.set_yticks([0,0.1,0.2,0.3,0.4])
ax.set_yticklabels([" ",0.1,0.2,0.3,0.4], fontsize = 20, fontname='Arial')
ax.set_xticks([el*2 for el in range(9)])
ax.set_xlim([-2,17])
ax.set_ylim([0,0.45])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)

plt.xticks(fontsize=24, fontname='Arial')
ax.tick_params(axis='x', which='major')
ax.tick_params(axis='y', which='major', direction='in',length=12, labelsize=24, width=2)

plt.tight_layout()
plt.savefig('Fig4_2_Cre-.jpg', dpi=600)
plt.savefig('Fig4_2_Cre-.eps', format='eps', dpi=600);
```
